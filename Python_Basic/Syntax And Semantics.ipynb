- Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structure programs in a language. In simpler Terms ,Syntax is about the correct arrangemsnts of words and symbols in a code 

- Semantics refers to the meaning of interpretation of the symbols,characters and commands in a language. it is about what the code is supposed to do when it runs.
