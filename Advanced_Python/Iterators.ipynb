### Iterators

##### Iterators  Are Advanced python concept that allows for efficient looping and memory management. Iterators provide a way to Access  element of a collection sequentially without exposing the underlined structure.

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
# Iterator
iterator=iter(my_list)
print(type(my_list))
```

```python
iterator
```

```python
## Iterate through all the elements
next(iterator)
```

```python
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```
