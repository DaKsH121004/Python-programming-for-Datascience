### Decorators

#### Decorators are a powerful and flexible feature and python that allows you to modify the behaviour of function or class method. They are commonly used to add functionality to the function or method without modifying their actual code.

```python
### Function copy 
## Closures function inside a function
```

```python
## Function Copy
def welcome():
    return " Hey Welcome here "
welcome()
```

```python
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
## Closures 
def main_welcome(msg):
    def sub_welcome_method():
        print("welcome you in this Repo")
        print(msg)

        print("please learn thses concepts ")
    return sub_welcome_method()
```

```python
main_welcome("hyy")
```

```python
def main_welcome(func):
    def sub_welcome_method():
        print("welcome you in this Repo")
        func("Myself Daksh")

        print("please learn thses concepts ")
    return sub_welcome_method()
```

```python
main_welcome(print)
```

```python
## Decorator
def main_welcome(func):
    def sub_welcome_method():
        print("welcome to this Repo")
        func()
        print("please learn these concepts ")
    return sub_welcome_method()
```

```python
def course_introduction():
    print("this is my github repo")

course_introduction()    
```

```python
main_welcome(course_introduction)
```

```python
@main_welcome
def course_introduction():
    print("this is my github repo")
```
