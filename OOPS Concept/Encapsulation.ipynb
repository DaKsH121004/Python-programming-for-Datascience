## Encapsulation And Abstraction

#### Encapsulation and Abstraction are two fundamental principle of object oriented programming(oops )that help and design robust maintainable and reusable Code. encapsulation involves Building data and methods that operates on data. Within a  single unit while abstraction involves hiding complex implementation details and exposing only the necessary features.

### Encapsulation

#### Encapsulation is the concept of wrapping  data (variables) and methods (functions) together as a single unit. It restricts direct access to someone of the object's component which is means of preventing accidental interface and misuse of data.

```python
## Relation with Getter and setter methods
## Public ,protected,private variables or access modifiers
class Person:
    def __init__(self, name, age):
        self.name = name      # public variable
        self.age = age        # public variable

    # Getter method
    def get_name(self):
        return self.name


person = Person("Daksh", 20)

print(person.name)       # direct access (public)
print(person.age)        # direct access (public)
print(person.get_name()) # access via getter
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name     # private variable
        self.__age = age       # private variable
        self.gender = gender  # public variable

    # Getter method
    def get_name(self):
        return self.__name


# Object creation (outside the class)
person = Person("Daksh", 20, "Male")

# Calling getter method
print(person.get_name())
```

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name      # protected variable
        self._age = age       # protected variable
        self.gender = gender  # public variable


class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)


employee = Employee("Krish", 34, "Male")
print(employee._name)
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.__account_holder = account_holder   # private variable
        self.__balance = balance                 # private variable

    # Getter method
    def get_balance(self):
        return self.__balance

    # Setter method
    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("Invalid balance amount")

    # Getter for name
    def get_account_holder(self):
        return self.__account_holder


# Object creation
account = BankAccount("Daksh", 5000)

# Access using getter
print("Account Holder:", account.get_account_holder())
print("Balance:", account.get_balance())

# Modify using setter
account.set_balance(7000)
print("Updated Balance:", account.get_balance())

# Invalid update
account.set_balance(-3000)
```
