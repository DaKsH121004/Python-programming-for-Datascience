- Exception is Pythonâ€™s built-in base class for all errors.
- Error is a user-defined base exception.
- dobException is a specific custom exception for age (DOB) validation.
- pass means no extra code is required; inheritance is enough.

```python
# Custom Exception Example

class Error(Exception):
    pass


class dobException(Error):
    pass


```

```python
year = int(input("Enter the dob: "))
age = 2026 - year

try:
    if age >= 20 and age <= 30:
        print("You can apply for the exam")
    else:
        raise dobException
except dobException:
    print("Sorry, your age should be greater than 20 or less than 30")
```
